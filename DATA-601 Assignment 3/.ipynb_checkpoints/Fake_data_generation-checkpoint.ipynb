Write a Python notebook that generates a file containing the following data:​ ​

Email addresses​. Must have an "@"
Phone numbers​
Home Address​
Person's name​
Year born. Use realistic values.​
Number of kids. Use realistic values.​
Categorical variable: rent or own?​
Annual income. Optional challenge: Use a non-uniform distribution​
Number of speeding tickets in past year. Optional challenge: Use a non-uniform distribution​
The user of your notebook should be able to specify how many entities are to be generated.​ ​

​Do not include the .csv output file in your submission -- the file should be generated dynamically.

Order of columns in CSV is not relevant

```python
import random
import string
import pandas as pd

def generate_emails():
    """
    This function returns random emails from the 
    domains listed.
    """
    letters = string.ascii_lowercase
    domain = ['@gmail.com', '@yahoo.com', '@outlook.com', '@umbc.edu']
    return "".join(random.sample(letters,(random.randrange(3,8))))+domain[(random.randrange(4))]

def generate_phonenumber():
    """
    This function returns random phonenumber
    from the given range
    """
    numbers = random.randint(1000000000,9999999999)
    return numbers

def generate_address():
    """
    This function returns random address using
    string and random functions
    """
    low_case = string.ascii_lowercase
    upp_case = string.ascii_uppercase
    return str(random.randint(10,9999))+' '+"".join(random.sample(low_case,(random.randrange(3,8))))+','+"".join(random.sample(upp_case,2))+' '+str(random.randint(10000,99999))

def generate_names():
    """
    This function returns names in uppercase
    letters
    """
    letters = string.ascii_uppercase
    return "".join(random.sample(letters,(random.randrange(4,9))))

def generate_year_born():
    """
    This fucntion returns random year from 
    range using random function
    """
    years = random.randint(1950,2000)
    return years

def generate_kids_num():
    """
    This function returns random number
    from given range using randint()
    """
    kids_num = random.randint(0,3)
    return kids_num

def generate_categroical_var():
    """
    This function return value from list
    using random function
    """
    category_var = ['Rent', 'Own']
    return category_var[(random.randrange(2))]

def generate_annual_income():
    """
    This function returns income from range
    '$' is concatenated before string
    """
    income = random.randint(30000,300000)
    return "$"+ str(income)

def generate_speeding():
    """
    This function returns number of speeding
    tickets generated using randint() in random
    function
    """
    tickets = random.randint(0,10)
    return tickets

def generate_fake_data(fake_data,number_of_entries):
    """
    This function takes 'file name' and 'number of entries'
    as input and returns '.csv' file of fake data
    """
    # 1) All the columns are initialized through a list
    df = pd.DataFrame(columns= ['Email id', 'Phone number', 'Address', 'person name', 'Number of Kids', 'Year born',\
                               'Categorical Variable', 'Annual income', 'Speeding tickets'])
    # 2) Loop for creating entries from given number
    for i in range(number_of_entries):
        df = df.append({'Email id': generate_emails(), 'Phone number': generate_phonenumber(), 'Address': generate_address(),\
                        'person name': generate_names(), 'Number of Kids': generate_kids_num(), 'Year born': generate_year_born(),\
                       'Categorical Variable': generate_categroical_var(), 'Annual income': generate_annual_income(),\
                       'Speeding tickets': generate_speeding()}, ignore_index= True)
    data_to_csv = df.to_csv('{}.csv'.format(fake_data)) # 3) to_csv is used to create .csv file from dataframe
    return df
    
        
n = int(input("Enter the required number of entries: ")) # Input for number of rows required
fake_data = str(input("Enter the .csv file name: ")) # Input required filename for fake data
generate_fake_data(fake_data,n)
```
