I provide you with a compressed XML file. Some of the fields contain HTML. ​

Extract the XML from the .zip file. In Python, use a module to parse the XML (do not write an XML parser)​
Using Python, extract the HTML from the XML. Then use BeautifulSoup4 to parse the HTML 
For each HTML page, report the number of links (URLs with the tag < a href="URL">text) in each HTML page 
Submit a single Jupyter notebook that parses the XML file and produces the count of links per HTML file.​

```python
from zipfile import ZipFile
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET

# Using ZipFile context manager to extract XML files from zip file
# Please enter the XML zip file path using '\\' instead of '\' for parsing.
with ZipFile(str(input('Please enter the XML zipfile path using \\\\ seperators: ')), 'r') as zipObj:
   zipObj.extractall()

# Since XML is hierarchial data format, so parsing the xml file using "ElementTree" module
tree = ET.parse('xml_containing_html.xml')
root = tree.getroot()

# Iterating through xml tags for HTML content
for child in root:
    author_name = child.get('author')
    for attr in child:
        if attr.tag=='content': # Searching for HTML content using XML tag "content"
            html_text = attr.text
            soup = BeautifulSoup(html_text,'html.parser') # Using beautifulsoup from bs4 to parse the html
            htmls = soup.find_all('a') # Finding all the URLs with 'a' tag using soup object
            print("The HTML file of 'author- {}' has URL links: {}".format(author_name,len(htmls)))
```
