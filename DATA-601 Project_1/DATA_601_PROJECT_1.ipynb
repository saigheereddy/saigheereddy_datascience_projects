## Exploratory Data Analysis using Jupyter and Python3 for data set 

### Data source location: https://opendata.maryland.gov/Public-Safety/Violent-Crime-Property-Crime-by-Municipality-2000-/2p5g-xrcb

### From the Exploratory Data Analysis of this data, we will get to know about the figures of all crimes in Maryland state happened between 1990 to 2017. Using this data, we can analyze the crime rates, type of crimes in various maryland state counties, Number of crimes happened in each county of maryland. This analysis helps in identifying the maryland counties which are more prone to crime.

```python
from io import StringIO
import pandas as pd
import matplotlib.pyplot as plt
import requests

# Using requests module to get data from source through URL
url = "https://opendata.maryland.gov/api/views/2p5g-xrcb/rows.csv?accessType=DOWNLOAD" # CSV source: https://opendata.maryland.gov/Public-Safety/Violent-Crime-Property-Crime-by-Municipality-2000-/2p5g-xrcb
s = requests.get(url).content # To read content from csv file
df = pd.read_csv(StringIO(s.decode('utf-8'))) # Decoding the utf-8 format
df
```

```python
# df.shape: To get number of rows and columns of the dataframe
df.shape
```

```python
# This is done to get each variable of the dataframe for implementation of EDA
# Below are the variables or columns of the .csv file
df.columns
```

```python
# isnull() function is used for cleaning the data.
# This results in showing number of null values in each variable or column
df.isnull().sum(axis = 0)
```

```python
# From the above, there are significant number of null values in the data.
# Each column with null values needs to be replaced with 'NaN' or '0'
# fillna() is used to replace the null values in each column

df['PERCENT CHANGE'].fillna('NaN', inplace=True)
df['VIOLENT CRIME PERCENT CHANGE'].fillna('NaN', inplace=True)
df['PROPERTY CRIME PERCENT CHANGE'].fillna('NaN', inplace=True)
df['OVERALL PERCENT CHANGE PER 100,000 PEOPLE'].fillna('NaN', inplace=True)
df['OVERALL CRIME RATE PER 100,000 PEOPLE'].fillna('NaN', inplace=True)
df['VIOLENT CRIME RATE PERCENT CHANGE PER 100,000 PEOPLE'].fillna('NaN', inplace=True)
df['PROPERTY CRIME RATE PERCENT CHANGE PER 100,000 PEOPLE'].fillna('NaN', inplace=True)
```

```python
# describe() is used to get basic statistical information
# Here 'GRAND TOTAL' variable includes values of total crimes happened in each year

print(df['GRAND TOTAL'].describe())
```

```python
# Grouping by year for average of total crimes happened in all counties of maryland state
list1=df.groupby(['YEAR'])['GRAND TOTAL'].mean()

# Using line plot for finding steep in increases and decreases
list1.plot(kind='line',figsize=(20,16),color='green',fontsize=20)
plt.suptitle('TOTAL CRIMES OF MARYLAND IN EACH YEAR', fontsize=20)
plt.xlabel('YEARS',fontsize=20)
plt.ylabel('COUNT',fontsize=20)
plt.show()

# Findings:
# -> From the figure we can analyze that, total number of crimes were highest around 1993-1995. 
# -> Starting from 1995, there is continuous decrease in total crimes around the state of maryland
```

```python
list1=df.groupby(['YEAR'])['GRAND TOTAL'].mean()

# Analyzing above information in bar graph helps in indentifying more accurate information
list1.plot(kind='bar',figsize=(20,16),color='blue',fontsize=20)
plt.suptitle('TOTAL CRIMES OF MARYLAND IN EACH YEAR', fontsize=20)
plt.xlabel('YEARS',fontsize=20)
plt.ylabel('COUNT',fontsize=20)
plt.show()

# From the figure below, we can analyze the following
# -> There is an increase in crime rate between 1990 to 1995.
# -> Although there is decrease in total crimes from 1995 to 2017, there are some fluctuations happened in 2000-2001, 2007-2008,
#    and slightly increased in 2017 compared to 2016.
# -> Maximum number of crimes were happpend in Maryland in 1995.
# -> From the data least number of crimes were happened in year 2016.
```

```python
# The total crime data is now grouped by violent and property crimes
list1=df.groupby(['YEAR'])['VIOLENT CRIME TOTAL', 'PROPERTY CRIME TOTALS'].mean()
list1.plot(kind='line',figsize=(20,10),fontsize=20)
plt.suptitle('VIOLENT AND PROPERTY CRIME FIGURES IN EACH YEAR', fontsize=20)
plt.xlabel('YEAR',fontsize=20)
plt.ylabel('COUNT',fontsize=20)
plt.show()

# From the line graph below, we can analyze the following:
# -> From the total crimes, property crimes are the most happening and violent crimes are the least.
# -> Property crimes are significantly reduced from 1995 to 2017, but violent crimes figures remains to be the same.
```

```python
list1=df.groupby(['YEAR'])['VIOLENT CRIME TOTAL', 'PROPERTY CRIME TOTALS'].mean()
list1.plot(kind='bar',figsize=(20,10),fontsize=20, stacked=True)
plt.suptitle('VIOLENT AND PROPERTY CRIME FIGURES IN EACH YEAR', fontsize=20)
plt.xlabel('YEAR',fontsize=20)
plt.ylabel('COUNT',fontsize=20)
plt.show()

# From this bar chart, we can see the following
# -> High average number of property crimes happened in 1995 and least in 2016
# -> High average number of violent crimes happened in 1992 and 1993, least in 2014.
```

```python
# Grouping average of each crime happened in maryland with year
list1=df.groupby(['YEAR'])['MURDER','RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT'].mean()
list1.plot(kind='bar',figsize=(20,10),fontsize=20, stacked=True)
plt.suptitle('STACKED FIGURES OF ALL CRIMES IN MARYLAND IN EACH YEAR', fontsize=20)
plt.xlabel('YEARS',fontsize=20)
plt.ylabel('COUNT',fontsize=20)
plt.show()

# From the stacked bar plot below, we can analyze the following.
# -> In all the years, "M/V THEFT" is the most happening crime all over the maryland.
# -> There is significiant decrease in average number of crimes of 'MURDER', 'M/V THEFT', 'LARCENY THEFT' & 'RAPE'
#    from 1990 to 2017
# -> Figures of average number of crimes including 'ROBBERY', 'B&E', 'AGG. ASSUALT', more or less remained same through out the
#    decades.
```

```python
import seaborn as sns
sns.set_style("whitegrid");

# Using scatter plot for finding the lowest and highest crime prone counties in Maryland
sns.FacetGrid(df, hue="COUNTY", height=10) \
   .map(plt.scatter, "GRAND TOTAL", "YEAR") \
   .add_legend();
plt.show();

# Findings from plot below:
# -> 'Baltimore City' county has the highest average crime rate in maryland.
```

```python
# Grouping by each county of maryland with different crimes
county_data=df.groupby(['COUNTY'])['MURDER','RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT'].mean()
county_data.plot(kind='bar',figsize=(20,10),fontsize=20, stacked=True)
plt.suptitle('ALL CRIME FIGURES IN EACH COUNTY OF MARYLAND', fontsize=20)
plt.xlabel('YEAR',fontsize=20)
plt.ylabel('COUNT',fontsize=20)
plt.show()

# From the bar graph below, we get the following information:
# -> Baltimore city has the highest average number of all crimes of all the counties in maryland
# -> Anne arundel, Allegany, Harford, Montogomery, Worcester, Wincomico etc
#    counties followed baltimore in highest average number of crimes.
```

```python
# Since 'BALTIMORE CITY' county showed highest average number of crimes of all the counties in maryland

# Groupby() is used to group all counties data to county_data object.
county_data=df.groupby('COUNTY')

# Getting baltimore city data using get_group() from country_data object
baltimore = county_data.get_group('Baltimore City')

# groupby() is used for grouping the average number of all crimes happened in baltimore in each year 
baltimore_crime = baltimore.groupby(['YEAR'])['MURDER','RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT'].mean()

# plot() is used to plot the line chart.
baltimore_crime.plot(kind='bar',figsize=(20,10),fontsize=20, stacked=True)
plt.suptitle('FIGURES OF EACH CRIME IN BALTIMORE CITY OVER YEARS 1990 - 2017', fontsize=20) #Figure title
plt.xlabel('YEAR',fontsize=20) # Naming xlabel with fontsize 
plt.ylabel('COUNT',fontsize=20) # Naming ylabel with fontsize
plt.show()


# Findings from the line chart below:
# -> Average crime rate of each crime in baltimore is significantly reduced in baltimore from 1995 to 2017
# -> 'Baltimore city' has more average number of 'LARCENY THEFT' crimes followed by 'B&E', 'ROBBERY', 'LARCENY THEFT',
#    'AGG. ASSAULT' crimes.
# -> 'MURDER' and 'RAPE' crimes have less average crime number in all the counties.
```

```python
baltimore_crime = baltimore.groupby(['YEAR'])['MURDER PER 100,000 PEOPLE', 'RAPE PER 100,000 PEOPLE',\
       'ROBBERY PER 100,000 PEOPLE', 'AGG. ASSAULT PER 100,000 PEOPLE',\
       'B & E PER 100,000 PEOPLE', 'LARCENY THEFT PER 100,000 PEOPLE',\
       'M/V THEFT PER 100,000 PEOPLE'].mean()
baltimore_crime.plot(kind='line',figsize=(20,10),fontsize=20)
plt.suptitle('BALTIMORE CITY EACH CRIME PER 100,000 PEOPLE OVER 1990 - 2017', fontsize=20)
plt.xlabel('YEAR',fontsize=20) # Naming xlabel with fontsize
plt.ylabel('COUNT',fontsize=20) # Naming ylabel with fontsize
plt.show()

# Findings from the stacked bar graph below:
# -> After 2016 we can see that there is an increase in average number of 'B&E', 'AGG. ASSAULT' & 'ROBBERY'. From these figures,
#    it is possible that there might be an increase in these crimes over the next few years.
```
